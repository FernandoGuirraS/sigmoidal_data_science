# Analisando Dados do COVID - 19 no Brasil e no Mundo

O ano de $2020$ foi responsável por uma grande mudança na forma de convivência da sociedade no mundo atual, ocasionada pela chegada do **Corona Vírus**, perturbando a o modo de vidas das pessoas no planeta.

O SARS-CoV-2, como é chamado cientificamente o novo coronavírus, teve o primeiro caso de Covid-19 identificado na cidade de Wuhan, na China, registrado em novembro. A abreviação SARS-CoV-2 vem das seguites abreviações:

* SARS é uma abreviação de uma Síndrome Respiratória Aguda Grave. Essa é a forma grave de muitas doenças respiratórias e o principal sintoma é a dificuldade de respirar;
* CoV é uma abreviação de coronavírus, a família de vírus que ele pertence;
* O número 2 é ao fato de ser parecido com uma outra espécie de coronavírus que quase virou uma pandemia em 2002, o SARS-CoV.

### Quais meses em que a taxa de reprodução do COVID-19 foi extrema no Brasil?

A variável `reproduction_rate` se refere taxa efetiva de reprodução (R) do COVID-19. O número $R$ representa o número, em média, de pessoas para as quais uma pessoa infectada transmitirá o vírus. Matematicamente, o número $R$ deve se encontrar abaixo de $1.0$ para que não haja propagação do número de casos. Se for superior a $1.0$, cada paciente transmite a doença a pelo menos mais uma pessoa, e o vírus se dissemina. Se é menor do que 1, cada vez menos indivíduos se infectam e o número dos contágios diminuem. Para maiores informações sobre a taxa de reprodução, simulações gráficas, pode encontrar no seguinte [site](https://www.bbc.com/news/health-52473523).

Segundos dados do *dataset*, o Brasil apresentou um pico no número de casos nos meses de março, abril e maio do ano de $2021$, chegando ao pico no dia $19$ de março de $2021$ atingindo uma taxa de reprodução de $2.47,$ ou seja,

> Cada $10$ pessoas o vírus era transmitido para $247$ diferentes indivíduos.

```python
# taxa de reproducao do covid no brasil
df.loc[df.location == 'Brazil', ['date','location', 'reproduction_rate']].sort_values(by='reproduction_rate',
                                                                                      ascending=False).reset_index(drop=True)[0:60: 6]
```

Com o avanços dos estudos, foram produzidas vacinas de diferentes farmacêuticas, que começaram a ser aplicadas nos indivíduos. O *dataset* mostra que a vacinação no **Brasil** iniciou no dia $17$ de Janeiro de $2021,$ onde um total de $112$ pessoas foram contempladas com uma das vacinas. Com os passar dos dias, o número de vacinados foi aumentando, e quando comparamos com o número de novos casos, novas mortes e a taxa de contaminação, podemos observar um aumento no início da vacinação, acompanhado de uma estabilização e seguido de queda. Em meados de *junho* a taxa de propagação do vírus encontrava-se abaixo de $1.0,$ onde, a partir daí, o número de casos e o número de mortes começaram a diminuir, mostrando eficácia da imunização.

## Visualização dos Dados

A etapa de visualização dos dados é uma das etapas mais importantes no desenvolvimento do projeto. Nela é realizada a representação dos dados de maneira visual, facilitando o entendimento através de figuras das informações dos dados disponíveis.

### Como está o número de casos no país que apresentou o primeiro caso?

O primeiro caso foi constatado na cidade de *Wuhan* na China. Houve um grande surto no período mas logo os casos foram contidos. O país apresentou várias medidas protetivas para conter a propagação do vírus que teve resultado rápido. O número de casos começou a apresentar queda e está estável até o momento.

### Qual a quantidade de testes de COVID - 19 são positivos no Estados Unidos?

O Estados Unidos (EUA) é um país que passou e passa por momentos difíceis na pandemia, devido ao crescente número de pessoas infectadas. Com o intuito de conter a propagação do vírus, uma das medidas adotadas foi testar em massa a população diariamente. No gráfico abaixo podemos observar uma grande quantidade de testes realizados diariamente, sendo que seu maior pico de testagem foi em torno de *janeiro de $2021$*, tendo como resultado um grande aumento nos testes com resultado positivo. Essa medida faz com que as pessoas infectadas sejam descobertas no início, para que possam passar por um período de incubação da doença. Isso fez com que os casos apresentassem uma queda.

```python
# criando dataframe com dados do estados unidos com numero de novos testes e a taxa positiva
df_tests = df.loc[df.location == 'United States', ['date', 'new_tests', 'positive_rate']]
df_tests.dropna(inplace=True)
df_tests.reset_index(drop=True, inplace=True)
#df_tests.set_index('date', inplace=True)
df_tests.head()

# transformando a taxa de novos testes para resultado real
positive_tests = []
for i in range(0, len(df_tests.new_tests)):
    aux = round(df_tests['new_tests'][i] * df_tests['positive_rate'][i], 0)
    positive_tests.append(aux)

df_tests['positive_tests'] = positive_tests
```

```python
# plotando numero de novos testes e testes positivos
fig, ax = plt.subplots(figsize=(12,8))

ax.plot(df_tests['date'], df_tests['new_tests'])
ax.fill_between(df_tests['date'], df_tests['new_tests'], alpha=0.5)
ax.plot(df_tests['date'], df_tests['positive_tests'])
ax.fill_between(df_tests['date'], df_tests['positive_tests'], alpha=0.5)
ax.set_title('Novos Testes de COVID - 19 x Pacientes com Resultado Positivo', size=16)

fig.autofmt_xdate()

plt.tight_layout();
```
